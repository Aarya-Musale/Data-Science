**Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
# We can perform diff types of analysis
#1. Univariate ---> We use single feature
#2. Bivariate  ---> We use 2 features
#3. Multi-variate ---> More than 2 features (State vs Population  ---> Gender)
```

```python
days = ['mon','tue','wed','thur','fri']
rain = [120,30,25,20,80]
```

```python
# Bi-variate plots = line plot,scatter plot,bar plot,barh plot

plt.plot(days,rain)
```

```python
plt.plot(days,rain,'--k8')
```

```python
# Always take independent features on x-axis
# and take dependent features on y-axis
```

```python
plt.plot(days,rain,'--k8')
plt.grid()
plt.title("Weekly Rainfall Data")
plt.xlabel("Day Mon-Fri")
plt.ylabel("Rainfall in mm")
```

```python
plt.plot(days,rain,'--k8',label = 'Daywise rainfall in mm')
plt.grid()
plt.title("Weekly Rainfall Data")
plt.xlabel("Day Mon-Fri")
plt.ylabel("Rainfall in mm")
plt.legend()   # This is used to make label visible used at frist line
plt.savefig('sample.png',dpi=400)
```

```python
# Bar plot

plt.bar(days,rain,color = ['r','m','c','k','y'],label = 'Daywise rainfall in mm')
plt.grid()
plt.title("Weekly Rainfall Data")
plt.xlabel("Day Mon-Fri")
plt.ylabel("Rainfall in mm")
plt.legend()
```

```python
# Barh plot

plt.barh(days,rain,color = ['r','m','c','k','y'],label = 'Daywise rainfall in mm')
plt.grid()
plt.title("Weekly Rainfall Data")
plt.ylabel("Day Mon-Fri")
plt.xlabel("Rainfall in mm")
plt.legend()
```

**Seaborn**

```python
# Seaborn directly works on DataFrame


import seaborn as sns
```

```python
# Load datasets

print(sns.get_dataset_names())
```

```python
p = sns.load_dataset('penguins')
p.head(2)
```

```python
p['species'].unique()
```

```python
# USing matplotlib

p['species'].value_counts()
```

```python
ps = p['species'].value_counts()
ps
```

```python
ps.index  # X-axis data
```

```python
ps.values   # Y-axis data
```

```python
plt.bar(ps.index,ps.values)
```

```python
sns.countplot(data=p,x='species')
```

```python
sns.countplot(data=p,x='island')
```
