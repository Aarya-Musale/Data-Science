**Numpy Operations**

```python
# Create an array of 4X4 zeros

import numpy as np

np.zeros((4,4))
```

```python
# create an array with constant value

np.full((4,4),12,dtype = float)
#np.full((4,4),0,dtype=bool)
```

```python
# Create a numpy array of 4X4 with diagonal elements as True and non-diagonal elements as False

np.identity(4)  #Square matrix
```

```python
#in python except 0 '' [] () None everything is True
np.identity(5,dtype=bool)
```

```python
# Create a numpy array of 4X4 with diagonal elements as False and non-diagonal elements as True
~np.identity(5,dtype=bool)
```

```python
# Access diagonal elements of numpy array

s = np.identity(5)
s.diagonal()
```

**Numpy Slicing and Indexing**

```python
x = np.arange(24).reshape((6,4))
x
```

```python
print(x[0])  #using index we can access one row at a time
print(x[3])
```

```python
x[3][3]   #row with element index

#we can also do like this
print(x[3,3])
```

```python
print(x[1,0])
print(x[5,3])
```

```python
x[2,2]
```

```python
# numpy array is mutable in nature
x[3,3]
```

```python
# replace 15 by 50
x[3,3] = 50
x
```

```python
#access multiple rows

x[:]  #[start_index:stop_index]

# When user does not give start and stop it gives you whole thing or whole matrix
```

```python
x[1:4]
```

```python
#access last 3 rows
x[3:]  # +ve indexing
x[-3:]  # -ve indexing
```

```python
x[::-3]
```

```python
x[::-2]
```

```python
# column operation

x[:,:]  #[row,column]
```

```python
x[-2:,-2:]
```

```python
x[2:4,1:3]
```

```python
x[2:4,::3]
```
