**Different ways of creating numpy arrays**

```python
# numpy array with random numbers

np.random.random(5)
```

```python
np.random.random((5,3))
```

```python
# 3D array
# (Block, Rows, Columns)

np.random.random((2,2,2))
```

```python
np.random
```

```python
# Generate array with sequence of numbers

c = np.arange(24)  # arange:array range
c
```

```python
c.size
```

```python
# Convert from 1D to 2D

# use reshape() which is Temporary

print(c.reshape((6,4)))
```

```python
print(c.reshape((6,2)))
```

```python
# use resize() which is permanant

c.resize((3,8))
```
