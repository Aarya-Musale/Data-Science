```python
# Assignment: try to create a series and DataFrame with str,tuple,list[list],list[tuple],Set,dict

# Series Using str
data = 'aarya','shivani'
s = pd.Series(data)
print(s)

# DataFrame using str
df = pd.DataFrame([['aarya'],['shivani'],['anshita'],['gatha']])
df
```

```python
# Series Using Tuple
tup_data = (1,2,5,'aarya')
s = pd.Series(tup_data)
print(s)

# DataFrame using Tuple
df_tup = (1,2,3), (4,5,6)
df = pd.DataFrame(df_tup)
print(df)

df = pd.DataFrame([[(1,2,3),(4,5,6)]])
print(df)

df = pd.DataFrame([(1,2,3),(4,5,6)])
df
```

```python
# Series Using list[list]
list_list = [[1,2,3],[4,5,6]]
s = pd.Series(list_list)
print(s)

# DataFrame using list
my_list = [1,2,3],[4,5,6]
df = pd.DataFrame(my_list)
print(df)

df = pd.DataFrame([[[1,2,3],[4,5,6]]])
df
```

```python
# Series Using list(tuple)

list_tuple = [(1,5,6),(4,3,2)]
s = pd.Series(list_tuple)
print(s)


# DataFrame Using list(tuple)
df = pd.DataFrame([(1,5,6),(4,3,2)])
print(df)

df = pd.DataFrame([[(1,5,6),(4,3,2)]])
print(df)
```

```python
# Series Using set

set1 = {10,20,30,40}
s = pd.Series(set1)
print(s)
```

```python
set1 = {10,20,30,40}
s = pd.Series(sorted(set1))
s

# DataFrame using set
set1 = {10,20,30,40}
df = pd.DataFrame(sorted(set1))
print(df)

df = pd.DataFrame([[{10,20,30,40}]])
df
```

```python
# Series using Dict
my_dict = {1:10,2:20,3:30}
s = pd.Series(my_dict)
print(s)

# DataFrame using dict
my_dict = {1:10,2:20,3:30}
df = pd.DataFrame(my_dict,index = [1])
print(df)

df = pd.DataFrame([[{1:10,2:20,3:30}]])
print(df)

df = pd.DataFrame([{1:10,2:20,3:30}])
print(df)

t = {'Num1':[1,2,3],'Num2':[4,5,6]}
pd.DataFrame(t)
t
```
